```python
from FocusTraining.focus_training import FocusLearning, StartToFocus
from FocusTraining.metrics import Accuracy
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import (Flatten, Dense, Input,
                                     Conv2D, MaxPool2D)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Model
from tensorflow.keras.models import load_model
import tensorflow.keras.backend as K
```

```python
pretrained_model = load_model('./models/pretrained_traditional/cifar10_1e_neg_4.h5',
                             compile=False)
```

```python
adam = Adam(1e-4) # Use same lr to avoid fine tuning argument
start_to_focus = StartToFocus()
```

```python
# model, dset_name, class_names, metrics, optimizer, start_to_focus
focus_learn = FocusLearning(pretrained_model, 'cifar10', list(range(10)),
                            [Accuracy()], adam, start_to_focus=start_to_focus)
```

```python
# n_epochs, batch_size=16, normal_grad_penalty=1, val_monitor_metric='loss', worst_k=2, focus_penalty=1e-2
train_history, total_val_history, class_val_history = focus_learn.train(n_epochs=100, batch_size=16, 
                                                                        normal_grad_penalty=1,
                                                                        val_monitor_metric='loss',
                                                                        worst_k=2, focus_penalty=1e-2)
```

```python
import pickle

# Save histories 
def save_history(path, data):
    with open(path, 'wb') as output:
        pickle.dump(data, output, pickle.HIGHEST_PROTOCOL)
        
save_history('./from_pretrained_train_history.pkl', train_history)
save_history('./from_pretrained_total_val_history.pkl', total_val_history)
save_history('./from_pretrained_class_val_history.pkl', class_val_history)
```

```python
# Read histories
def read_history(path):
    with open(path, 'rb') as input:
        data = pickle.load(input)
        
    return data
```
